1. Relational Plot 
- to see the statestical relation between 2 or more variable
-Bivariate Analysis
Plots
- scatter plot 
- line plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
# scatter plot => axes level function (hue it change catagories)
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# replot => figure level function =>squre shape
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap=px.data.gapminder()
temp=gap[gap['country']=='India']
temp
```

```python
# axes level function
sns.lineplot(data=temp,x='year',y='lifeExp')
```

```python
#using relplot
sns.relplot(data=temp,x='year',y='lifeExp',kind='line')
```

```python
temp1=gap[gap['country'].isin(['India','Brazil','Germany','China'])]
temp1
```

```python
sns.relplot(data=temp1,x='year',y='lifeExp',kind='line',hue='country',style='continent')
```

```python
# facet plot => only work with figure level plot
sns.relplot(kind='scatter',data=tips,x='total_bill',y='tip',col='sex',row='smoker')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

2. Distribution Plot
- used for univariate analysis
- used to find out the distribution 
- range of the observation
- central tendancy
- is the data bimodel?
- are there outliers?

Plots
- Histogram
- kdeplot
- rugplot

```python
# figure level function -> diskplot
# axes level -> histplot, kdeplot,rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill')
```

```python
# bins
sns.displot(data=tips,x='total_bill',bins=10)
```

```python
# it's also possible to visualize the distribution of a categorical variable usinf the logic of a histogram
# Discrete bins are automatically set for categot=rical variable

#like count plot
sns.displot(data=tips,x='day',hue='time',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python

sns.displot(data=titanic,x='age',hue='sex',element='step')
```

```python
sns.displot(data=titanic,x='age',col='sex')
```

```python
# kdeplot
# rather than using discrete bins , a KDE plot smoots the observation with a gaussion kernel, producing a continous density estimates
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(kind='kde',data=tips,x='total_bill',hue='sex',fill='True')
```

```python
# in figure level function we cant change  size by matplotlib
# it change by passing hight and aspecct for manage with ratio
sns.displot(kind='kde',data=tips,x='total_bill',hue='sex',fill='True',height=10,aspect=1)
```

```python
# rug plot
# plot marginal distribution by drawing tricks along the x and y axes

# this function is intended to complete other plotsby showing the location of individual observation in an unbrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# a bivariate histogram bins the data within rectangle that tile the plot
# and then show the count of obeservation within each rectangle with the fill order
#sns.histplot(data=tips,x='total_bill',y='tip')
sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate Kdeplot smooths the (x,y) observation with a 2D Gausian
sns.displot(data=tips,x='total_bill',y='tip',kind='kde')
```

4. Matrix Plot


plots
- heatmap
- clustermap

```python
# heatmap
# plot rectangular data as a color encoded matrix
h=gap.pivot(index='country',columns='year',values='lifeExp')
# axes level function (it does not have figure finction)
plt.figure(figsize=(15,15))
sns.heatmap(h)
```

```python
plt.figure(figsize=(15,15))
eu=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
sns.heatmap(eu,annot=True,linewidths=0.5,cmap='viridis')
# annot shares values & linewidth create gap between boxes
```

```python
# clustermap
# plot  a matrix dataset as a hierarchical- clustered heatmap
# this function requires scipy to be available.
iris=px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
