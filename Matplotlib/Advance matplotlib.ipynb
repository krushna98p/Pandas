```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=pd.read_csv('iris.csv')
iris.head()
iris['Species'].sample(5)
```

```python
iris['Species'].sample(5)
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

### coloured scatter plot

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='summer',alpha=0.5) # alpha is opacity of grapn, cmap is color
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
```

```python
# plot size
plt.figure(figsize=(10,10))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.colorbar()
```

### Annotation

```python
batter=pd.read_csv('batter.csv')
df=batter.sample(25)
df
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
```

### Horizontal and vertical lines 

```python
plt.figure(figsize=(10,10))
plt.scatter(df['avg'],df['strike_rate'],s=df['runs'])
for i in range(df.shape[0]):
    plt.text(df['avg'].values[i],df['strike_rate'].values[i],df['batter'].values[i])
plt.axhline(130, color='red')
plt.axvline(30,color='red')
```

### subplot

```python
# a diss way to plot graph
plt.scatter(batter['avg'],batter['strike_rate'])
```

```python
# 2nd way
fig , ax = plt.subplots(figsize=(10,10))
ax.scatter(batter['avg'],batter['strike_rate'])
ax.set_title('avg vs strike rate')
ax.set_xlabel('avg')
fig.show()
```

```python
# 2 sise by side graph
fig ,ax =plt.subplots(nrows=2,ncols=1,sharex=True)
ax[0].scatter(batter['avg'],batter['strike_rate'])
ax[1].scatter(batter['avg'],batter['runs'])
ax[0].set_title('avg vs stike rate')
ax[1].set_title('avg vs runs')
ax[1].set_xlabel('avg')
ax[0].set_ylabel('strike rate')
ax[1].set_ylabel('runs')
```

```python
fig ,ax =plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batter['avg'],batter['strike_rate'])
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['strike_rate'])

ax[0,0].set_title('avg vs stike rate')
ax[0,1].set_title('avg vs runs')
ax[0,1].set_xlabel('avg')
ax[0,0].set_ylabel('strike rate')
ax[0,1].set_ylabel('runs')
ax[1,0].set_title('avg hist')
ax[1,1].set_title('strike rates hist')
```

```python
# 2nd way
fig =plt.figure()
ax1=fig.add_subplot(221)
ax1.scatter(batter['avg'],batter['strike_rate'])
ax2=fig.add_subplot(222)
ax2.scatter(batter['avg'],batter['runs'])
ax3=fig.add_subplot(223)
ax3.hist(batter['avg'])
ax4=fig.add_subplot(224)
ax4.hist(batter['strike_rate'])
fig.show()
```

### 3D Surface Plot

```python

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
xx ,yy=np.meshgrid(x,y)
z=xx**2+yy**2
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z=np.sin(xx) + np.cos(yy)
fig=plt.figure()
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
#scatter + surface together

fig=plt.figure()
ax=plt.subplot()
p=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python

fig=plt.figure()
ax=plt.subplot()
p=ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Heatmap

```python
deli=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
deli
```

```python
temp=deli[(deli['ballnumber'].isin([1,2,3,4,5,6]))& (deli['batsman_run']==6)]
temp
```

```python
grid=temp.pivot_table(index=deli['overs'],columns=deli['ballnumber'],values=['batsman_run'],aggfunc='count')
grid
```

```python
plt.figure(figsize=(10,10))
plt.imshow(grid)
plt.yticks(deli['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
```
