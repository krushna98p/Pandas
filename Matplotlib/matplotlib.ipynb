### 2D line plot

<li>Bivariate Analysis</li>
<li>categorical => numerial and numercal => numerical</li>
<li>usecase time series data</li>

```python
price=[20000,30000,43000,43000]
year=[2021,2022,2023,2024]
plt.plot(year,price)
```

```python
sk=pd.read_csv('sharma-kohli.csv')
sk.head()
```

```python
plt.plot(sk['index'],sk['V Kohli'])
```

```python
# ploting multiple plots
plt.plot(sk['index'],sk['V Kohli'])
plt.plot(sk['index'],sk['RG Sharma'])
```

```python
# labels title

plt.plot(sk['index'],sk['V Kohli'])
plt.plot(sk['index'],sk['RG Sharma'])
plt.title('virat and rohot compre')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
# color(hex) and line(width and style) and marker (size)
plt.plot(sk['index'],sk['V Kohli'],color='red', linestyle='dashed',linewidth=3,marker='D',markersize=3)
plt.plot(sk['index'],sk['RG Sharma'],color='black',linestyle='dotted',linewidth=2,marker='+',markersize='2')
plt.title('virat and rohot compre')
plt.xlabel('season')
plt.ylabel('runs')
```

```python
#legend location
plt.plot(sk['index'],sk['V Kohli'],color='red',label='virat')
plt.plot(sk['index'],sk['RG Sharma'],color='black',label='rohit')
plt.title('virat and rohot compre')
plt.xlabel('season')
plt.ylabel('runs')
plt.legend()# we can pass here location loc='___'
```

```python
#limiting axes
price=[20000,30000,43000,43000,4500000]
year=[2021,2022,2023,2024,2025]
plt.plot(year,price)
plt.ylim(0,100000)
```

```python
# grid & show
plt.plot(sk['index'],sk['V Kohli'],color='red',label='virat')
plt.plot(sk['index'],sk['RG Sharma'],color='black',label='rohit')
plt.title('virat and rohot compre')
plt.xlabel('season')
plt.ylabel('runs')
plt.legend()
plt.grid()
plt.show()
```

### Scatter plot

<li> Bivariate Analysis</li>
<li> numerical vs numerical </li>
<li> use case - finding correlation</li>

```python
# plt.scatter simple function
x= np.linspace(-10,10,50)
y=10*x+3 + np.random.randint(0,300,50)
plt.scatter(x,y)
```

```python
batter=pd.read_csv('batter.csv')
df=batter.head(50)
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.title('avg vs strike rate')
plt.xlabel('avg')
plt.ylabel('strike rate')
```

```python
# marker
plt.scatter(df['avg'],df['strike_rate'],marker='+')
```

```python
# size 
tips=sns.load_dataset('tips')
tips

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# scatterplot using plt.plot
plt.plot(tips['total_bill'],tips['tip'],'o')
# this was faster but you cant maupulate color 
```

### Bar chart

<li> Bivariate Analysis</li>
<li> Numerical vs Categorical </li>
<li> use casse - aggerecate analysis of groups</li>

```python
child=[1,2,3,4]
col=['red','green','blue','black']
plt.bar(col,child)
```

```python
# horizontal bar chart
plt.barh(col,child)
```

```python
season=pd.read_csv('batsman_season_record.csv')
season
```

```python
# multiple bar chart 
plt.bar(np.arange(season.shape[0])-0.3,season['2015'],width= 0.3)
plt.bar(np.arange(season.shape[0]),season['2016'],width= 0.3)
plt.bar(np.arange(season.shape[0])+0.3,season['2017'],width= 0.3)
plt.xticks(np.arange(season.shape[0]),season['batsman'])
plt.show()
```

```python
# problem
child=[1,2,3,4]
col=['red red red red','green green green green ','blue blue blue blue ','black']
plt.bar(col,child)

# solution
plt.xticks(rotation='vertical')
```

```python
#stacked bar chart
plt.bar(season['batsman'],season['2015'],label='2015')
plt.bar(season['batsman'],season['2016'],bottom=season['2015'],label='2016')
plt.bar(season['batsman'],season['2015'],bottom=(season['2015']+season['2016']),label='2017')
plt.legend()
```

### Pie chart

<li> Univariate & Bivariate Analysis</li>
<li> categorical vs numerical</li>
<li> use case -to finnd contribution on a standard scale</li>

```python
plt.pie(data)
subject=['math','sci','hindi','phy','chem']
plt.show()
```

```python
plt.pie(data,labels=subject)
```

```python
# percentage
cg=pd.read_csv('gayle-175.csv')
cg
plt.pie(cg['batsman_runs'],labels=cg['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# explode & shadow
plt.pie(cg['batsman_runs'],labels=cg['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```
