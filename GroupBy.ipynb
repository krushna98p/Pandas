# Group By

```python
import numpy as np
import pandas as pd
```

```python
deli=pd.read_csv("deliveries.csv")
imbd=pd.read_csv("imdb-top-1000.csv")
ipl=pd.read_csv("ipl-matches.csv")
imbd
```

```python
# Groupby
# it forms group on columns on the basis of categorial data
genre=imbd.groupby("Genre")
```

```python
# using aggrigate function on groupby
genre.max()
```

```python
# find top 3 genre by total earning
imbd.groupby("Genre").sum()['Gross'].sort_values(ascending=False).head(3)
```

```python
imbd.groupby("Genre")['Gross'].sum().sort_values(ascending=False).head(3)
```

```python
# find the genre whighst avg imbd rating
imbd.groupby("Genre")['IMDB_Rating'].mean().sort_values(ascending=False).head()
```

```python
# find director whih most popularity
imbd.groupby('Director')['No_of_Votes'].sum().sort_values(ascending=False).head()
```

```python
# highest rated movie of each genre
imbd.groupby('Genre')['IMDB_Rating'].max()
```

```python
# find the number of movies done by each actor
imbd.groupby('Star1')['Series_Title'].count().sort_values(ascending=False).head()
```

```python
# groupbby attributes and methods 
# find total number of groups ==> len
# find items in each group ==>size
# first()/last() => nth item
# get_group => vs filtering
# groups
# describe
# sample
# nunique
```

```python
# agg method
# passing dict
genre.agg({
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'mean'

})
```

```python
genre.agg({
    'Runtime':['mean','max'],
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'mean'

})
```

```python
# looping on groups
df=pd.DataFrame(columns=imbd.columns)
for group,data in genre:
    print(data[data['IMDB_Rating']==data['IMDB_Rating'].max()])
    
```

```python
# split(apply) combine
# apply => builtin function
genre.apply(min)
```

```python
# find the number of movies starting with A for each group
def a(group):
    return group['Series_Title'].str.startswith("A").sum()
genre.apply(a)
```

```python
# groupby on multiple cols
# size
# get_group
dio=imbd.groupby(['Director','Star1'])
dio.size()
```

```python
# find the most earning actor => director combo
dio['Gross'].sum().sort_values(ascending=False)
```

```python
imbd.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False)
```

```python
imbd.set_index('Series_Title',inplace=True)
```

```python
# agg on multiple groupby
dio.agg(['min', 'max'])
```
