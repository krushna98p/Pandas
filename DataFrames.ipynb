```python
import numpy as np
import pandas as pd
```

# DataFrames

```python
ipl=pd.read_csv('ipl-matches.csv')
movies=pd.read_csv("movies.csv")
runs=pd.read_csv("batsman_runs_ipl.csv")
ipl.head()
```

```python
# value_count
# player who won most player of the matech not normal match
ipl[~ipl['MatchNumber'].str.isdigit()]['Player_of_Match'].value_counts()
```

```python
# pie chart of toss decision
ipl['TossDecision'].value_counts().plot(kind='pie')
```

```python
# how many matches played by teams
(ipl['Team1'].value_counts() + ipl['Team2'].value_counts()).sort_values(ascending=False)
```

```python
# sort_values
# it sorts values acsending or desending
# it also sort values with nan vaalues by passing na_position you can get on top
movies.sort_values('title_x')
```

```python
movies.sort_values(['year_of_release','title_x'],ascending=[True,False])
```

```python
# rank (series)
runs['rank']=runs['batsman_run'].rank(ascending=False)
runs.sort_values('rank')
```

```python
# sort_index
# it sort index 
movies.sort_index(ascending=False)
```

```python
# set_index
# its sets column ti index

movies.set_index('title_x',inplace=True)
```
