```python
titanic=pd.read_csv('titanic.csv')
titanic['Name']
```

```python
# common function
# lower/higher/capitalize/title
titanic['Name'].str.lower()
#len
titanic['Name'][titanic['Name'].str.len()==82].values[0]
# strip
titanic['Name'].str.strip()
```

```python
# split =>get
titanic['last_name']=titanic['Name'].str.split(',').str.get(0)
titanic.head()
```

```python
titanic [['title','first_name']]=titanic['Name'].str.split(',').str.get(1).str.strip().str.split(' ',n=1,expand=True)
titanic.head()
```

```python
titanic['title'].value_counts()
```

```python
# replace
titanic['title']=titanic['title'].str.replace('Ms.','Miss.')
titanic['title']=titanic['title'].str.replace('Mlle.','Miss.')
```

```python
titanic['title'].value_counts()
```

```python
# filtering
# startswith/endswith
titanic[titanic['first_name'].str.startswith('A')]
# isdigit/isalpha
titanic[titanic['Name'].str.isdigit()]
```

```python
# apply ragex
#contain
# search john => both case
titanic[titanic['first_name'].str.contains('john',case=False)]
```

```python
# find lastname starts eith vowel and end with vowel
titanic[titanic['last_name'].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]
```

# datetime
### Timestamp object
time stamps reference particular moment in time (e.g., oct 24th,2022 at 9:00 pm)

```python
# creating a timestamp
pd.Timestamp('2025/4/2')
```

```python
# variation
pd.Timestamp('2025-4-2')
pd.Timestamp('2025, 4, 2')
```

```python
# only year
pd.Timestamp('2025')
```

```python
# using text
pd.Timestamp('2nd April 2025')
```

```python
pd.Timestamp('2025-4-2-9:23PM')
```

```python
# using datetime .datetime object in python
import datetime as dt
x=pd.Timestamp(dt.datetime(2025, 4, 3, 9 ,44))
```

```python
# fatching attributes
x.year
x.month
x.day
x.hour
x.minute
```

```python
# why seperates object to handle data and time when python already has datetime function
```

1. syntax wise datetime is very convenient
2. but the performance takes a hit while working with huge data. List vs Numpy Array
3. the weakness of python's datetime format inspired the Numpy team to add a set of native time series data type to Numpt.
4. the datetime64 dtype encodes dates as 64-bit integers, and thus allows of dates to be represented very compactly

```python
date=np.array('2025-04-03', dtype=np.datetime64)
date
```

```python
# vectorized operation
date + np.arange(12)
date
```

1. because of the uniform type in Numpy datetime64 arrays this type of operation can be accomplished much more quickly than if we were working directly with python's datetime objects, especially as arrays get large
2. pandas timestampobject combines the ease-ofuse of python datetime with the effifient storage and vectorize interface of numpy.datetime64
3. from a group of these timestamp object, pandas can construct a datetimeindex that can be used to index in a series or dataframe 

### DatetimeIndex object

a collection of pandas timestamp

```python
pd.DatetimeIndex(['2025-3-3','2023-4-4','2025-5-5'])
```

```python
# using python datetime object
pd.DatetimeIndex([dt.datetime(2023,2,3),dt.datetime(2023,2,3),dt.datetime(2023,2,3)])
```

```python
# from timestamp
dtin=pd.DatetimeIndex([pd.Timestamp(2023,2,23),pd.Timestamp(2023,2,24),pd.Timestamp(2023,2,25)])
```

```python
# using datetime insex as series
pd.Series([1,2,3],index=dtin)
```
